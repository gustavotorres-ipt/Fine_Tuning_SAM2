```python
import os
import random
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
geobodies_files = os.listdir("inputs") 
 
random.shuffle(geobodies_files)
```

```python
for filename in geobodies_files[:10]:

    image_path = f"./inputs/{filename}"
    label_path = f"./labels/{filename}"

    image = np.array(Image.open(image_path))
    image = np.stack([image, image, image], axis=2)

    label = np.array(Image.open(label_path)) * 255
    label = np.stack([label, label, label], axis=2)

    plt.axis("off")
    plt.imshow(np.hstack((image, label)))
    plt.show()
```
